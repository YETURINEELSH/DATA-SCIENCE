```python
import numpy as np
from scipy.spatial import distance
pointA=np.array([2,3,4])
pontB=np.array([5,6,7])
#Euclidean Distance
euclidean_distance=distance.euclidean(pointA,pontB)
print("Euclidean distance:",euclidean_distance)

similarity_distance=1/(1+euclidean_distance)
print("Euclidean similarity:",similarity_distance)
#Manhattan Distance
manhattan_distance=distance.cityblock(pointA,pontB)
print("Manhattan distance:",manhattan_distance)

similarity_distance=1/(1+manhattan_distance)
print("Manhattan similarity:",similarity_distance)
#Minkowski distance with p=3
minkowski_distance=distance.minkowski(pointA,pontB,p=3)
print("Minkowski distance(p=3):",minkowski_distance)

similarity_distance=1/(1+minkowski_distance)
print("Minkowski similarity(p=3):",similarity_distance)
#Minkowski distance with p=2
minkowski_distance=distance.minkowski(pointA,pontB,p=2)
print("Minkowski distance(p=2):",minkowski_distance)

similarity_distance=1/(1+minkowski_distance)
print("Minkowski similarity(p=2):",similarity_distance)
#Minkowski distance with p=1
minkowski_distance=distance.minkowski(pointA,pontB,p=1)
print("Minkowski distance(p=1):",minkowski_distance)

similarity_distance=1/(1+minkowski_distance)
print("Minkowski similarity(p=1):",similarity_distance)

```

```python
#Comparing two strings of same length using HAMMING METHOD
def hamming_distance(str1,str2):
  if len(s1)!=len(s2):
    raise ValueError("String must be of equal Length")
  return sum(ch1!=ch2 for ch1,ch2 in zip(s1,s2))
s1="Neelesh"
s2="NNNNNNN"
distance=hamming_distance(s1,s2)
print(f"Hamming Distance between '{s1}' and '{s2}' : {distance}")
```

```python
#JACCARD INDEX which is used to check the intersection over union of sets
#COSINE measures angle btw two vectors
def jaccard_index(str1,str2):
    set1,set2=set(str1.split()),set(str2.split())
    intersection=set1.intersection(set2)
    union=set1.union(set2)
    return len(intersection),len(union)
s1="data science is fun"
s2="science makes data useful"
print("Jaccard index:",*jaccard_index(s1,s2))
```

```python
#COSINE SIMILARITY- finding angle btw two vectors
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
s1="data science is fun"
s2="science makes data useful"
vectorizer=CountVectorizer().fit([s1,s2])
vectors=vectorizer.transform([s1,s2])
cos_sim=cosine_similarity(vectors[0],vectors[1])[0][0]
print("Cosine similarity:",cos_sim)
```

```python
# Sequence Based Measures
def lcs_length(X, Y):
    m, n = len(X), len(Y)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m):
        for j in range(n):
            if X[i] == Y[j]:
                dp[i + 1][j + 1] = dp[i][j] + 1
            else:
                dp[i + 1][j + 1] = max(dp[i][j + 1], dp[i + 1][j])

    return dp[m][n]

seq1 = "ABCDEF"
seq2 = "ABCDEG"

length = lcs_length(seq1, seq2)
print(f"Longest Common Subsequence Length between '{seq1}' and '{seq2}': {length}")
```
