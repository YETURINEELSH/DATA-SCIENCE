```python
import matplotlib.pyplot as plt
y1=[]
y2=[]
x=range(-100,100,10)
for i in x:
  y1.append(i**2)
for i in x:
  y2.append(-i**2)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('simple graph')
plt.axhline(color='black')
plt.axvline(color='black')
plt.show()
```

```python
import matplotlib.pyplot as plt
y1=[]
y2=[]
x=range(-100,100,10)
for i in x:
  y1.append(i**2)
for i in x:
  y2.append(-i**2)
plt.plot(x,y1,color='red')
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('simple graph')
plt.axhline(color='black')
plt.axvline(color='black')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('temporal.csv')

plt.plot(df['Mes'],df['deep learning'], label='deep learning')
plt.plot(df['Mes'],df['machine learning'], label='machine learning')
plt.plot(df['Mes'],df['data science'],label='data science' )

plt.xlabel('Data')
plt.ylabel('Popularity')
plt.title('Popularity of AI terms by data')

plt.legend()
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Line plot
import matplotlib.pyplot as plt

x_values = [0, 1, 2, 3, 4, 5]
y_values = [0, 1, 4, 9, 16, 25]

plt.plot(x_values, y_values)

plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Line Plot')

plt.grid(True)
plt.show()
```

```python
#Scatter plot
import matplotlib.pyplot as plt
x_values=[0,1,2,3,4,5]
y_values=[0,1,4,9,16,25]
plt.scatter(x_values,y_values)
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid()
plt.show()
```

```python
#Scatter plot
import matplotlib.pyplot as plt
x_values=[0,1,2,3,4,5]
y_values=[0,1,4,9,16,25]
plt.scatter(x_values,y_values,label='s',color='red',marker='^',s=30)
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid()
plt.show()
```

```python
#Bar plot
import matplotlib.pyplot as plt
values=[5,6,3,7,2]
names=['A','B','C','D','E']
plt.bar(names,values)
plt.title("Bar Graph")
plt.xlabel("Values")
plt.ylabel("Names")
plt.show()
```

```python
#bar horizontak plot
import matplotlib.pyplot as plt
values=[5,6,3,7,2]
names=['A','B','C','D','E']
plt.barh(names,values,color='red')
plt.title("Bar Graph")
plt.xlabel("Values")
plt.ylabel("Names")
plt.show()
```
